```r
#HOMEWORK2-TASK1-PART a

data = read.csv("C:/Users/negin/Desktop/Musk1.csv",sep=';',header = FALSE)
colnames(data) = c("Bag_class", "Bag_id", c(paste("Feature", 1:(ncol(data)-2), sep="")))
```

```r
data1=data[,3:168]
```

```r
#PCA
pca=princomp(data1, cor = TRUE, scores = TRUE)
plot(pca)
```

```r
str(pca)
```

```r
summary(pca)
```

```r
#component1  covers 31.18% of the variance,component2 together covers 13.92% of the variance and so on, Which means that component1
#is really small and we can not use it to show the information. The data is not divided in to labeled classes in a good way.
```

```r
pca.var= pca$sdev^2 
pca.var = round(pca.var/sum(pca.var)*100,2)
```

```r

par(mfrow=c(1,2))

plot(pca.var , xlab =" Principal Component ", ylab =" Proportion of Variance ")
plot(cumsum (pca.var), xlab =" Principal Component ", ylab ="Cumulative Proportion of Variance ")
```

```r
data$color[data$Bag_class == 0] = "blue" 
data$color[data$Bag_class == 1] = "red" 
plot(pca$scores[,1], pca$scores[,2], col=data$color,
     xlab = paste("PCA1 - ", pca.var.per[1], "%", sep=""),
     ylab = paste("PCA2 - ", pca.var.per[2], "%", sep=""),
     main = "PCA")
```

```r
#MDS
dist_matrix = dist(scale(data1, center=TRUE, scale=TRUE), method="euclidean")
MDS=cmdscale(dist_matrix, eig=TRUE)
mds.var = round(MDS$eig/sum(MDS$eig)*100,2)
par(mfrow=c(1,2))
plot(mds.var, main = "Scree Plot", xlab = "MDS", ylab = "Percent of Variation")
plot(cumsum(mds.var), main = "Cumulative Scree Plot",
     xlab = "MDS",ylab = "Cumulative Percent of Variance")
```

```r
mds$values = mds$points
plot(mds$values[,1], mds$values[,2], col=data$color,
     xlab = paste("MDS1 - ", mds.var[1], "%", sep=""),
     ylab = paste("MDS2 - ", mds.var[2], "%", sep=""),
     main = " MDS ")
```

```r
#From the PCA and MDS two dimensional figures we can understand that the results for both methods are similar. if we 
#use two components it will cover approximately 45% of the total variance. which is not very high value. It means that we have to 
#use approximately 35 components to cover approximately 95 percent of  the variance.
```

```r
img=readJPEG("C:/Users/negin/Desktop/Task2.jpg")
str(img)
```

```r
par(mfrow=c(1,1))
plot(0:256,0:256,type="n")
rasterImage(img,0,0, 256,256,interpolate=FALSE)
```

```r
#Three channels
red=img[1:256,1:256,1]
green=img[1:256,1:256,2]
blue=img[1:256,1:256,3]
```

```r
#Adding noise to image
red_noise=matrix(runif(256*256,min(red),0.1*max(red)),256,256)
green_noise=matrix(runif(256*256,min(green),0.1*max(green)),256,256)
blue_noise=matrix(runif(256*256,min(blue),0.1*max(blue)),256,256)
red=red+red_noise
green=green+green_noise
blue=blue+blue_noise
img[1:256,1:256,1]=red
img[1:256,1:256,2]=green
img[1:256,1:256,3]=blue
img=ifelse(img>1,1,img)
plot(0:256,0:256,type="n")
rasterImage(img,0,0,256,256,interpolate=FALSE)
```

```r
#displaying each channel separately using image function
par(mfrow=c(1,3))
image(1:256,1:256,t(apply(red,2,rev)))
image(1:256,1:256,t(apply(green,2,rev)))
image(1:256,1:256,t(apply(blue,2,rev)))
img_red=img
img_red[,,2:3]=0
plot(0:256,0:256,type="n")
rasterImage(img_red,0,0,256,256,interpolate=FALSE)
img_green=img
img_green[,,c(1,3)]=0
plot(0:256,0:256,type="n")
rasterImage(img_green,0,0,256,256,interpolate=FALSE)
img_blue=img
img_blue[,,1:2]=0
plot(0:256,0:256,type="n")
rasterImage(img_blue,0,0,256,256,interpolate=FALSE)
```

```r
#Transform noisy image to grayscale image
img_grayscale=img[,,1]+img[,,2]+img[,,3]
img_grayscale=img_grayscale/max(img_grayscale)
plot(c(0,1),c(0,1),t='n')
rasterImage(img_grayscale,0,0,1,1)
```

```r
#data matrix
patches_matrix=matrix(0,53824,625)
i=1
for (n in 13:244){
    for (k in 13:244){
        patch1=img_grayscale[n-12,(k-12):(k+12)]
        patch2=img_grayscale[n-11,(k-12):(k+12)]
        patch3=img_grayscale[n-10,(k-12):(k+12)]
        patch4=img_grayscale[n-9,(k-12):(k+12)]
        patch5=img_grayscale[n-8,(k-12):(k+12)]
        patch6=img_grayscale[n-7,(k-12):(k+12)]
        patch7=img_grayscale[n-6,(k-12):(k+12)]
        patch8=img_grayscale[n-5,(k-12):(k+12)]
        patch9=img_grayscale[n-4,(k-12):(k+12)]
        patch10=img_grayscale[n-3,(k-12):(k+12)]
        patch11=img_grayscale[n-2,(k-12):(k+12)]
        patch12=img_grayscale[n-1,(k-12):(k+12)]
        patch13=img_grayscale[n,(k-12):(k+12)]
        patch14=img_grayscale[n+1,(k-12):(k+12)]
        patch15=img_grayscale[n+2,(k-12):(k+12)]
        patch16=img_grayscale[n+3,(k-12):(k+12)]
        patch17=img_grayscale[n+4,(k-12):(k+12)]
        patch18=img_grayscale[n+5,(k-12):(k+12)]
        patch19=img_grayscale[n+6,(k-12):(k+12)]
        patch20=img_grayscale[n+7,(k-12):(k+12)]
        patch21=img_grayscale[n+8,(k-12):(k+12)]
        patch22=img_grayscale[n+9,(k-12):(k+12)]
        patch23=img_grayscale[n+10,(k-12):(k+12)]
        patch24=img_grayscale[n+11,(k-12):(k+12)]
        patch25=img_grayscale[n+12,(k-12):(k+12)]
        patches_matrix[i,]=c(patch1,patch2,patch3,patch4,patch5,patch6,patch7,patch8,patch9,patch10,patch11,patch12,patch13,patch14
                            ,patch15,patch16,patch17,patch18,patch19,patch20,patch21,patch22,patch23,patch24,patch25)
    i=i+1}}
#PCA
pca_image=princomp(patches_matrix, cor = TRUE, scores = TRUE)
plot(pca_image)
```

```r
str(pca_image)
```

```r
summary(pca_image)
```

```r
#We can see that component1 covers  69% of the total variance and the component2 covers 6% of the total variance 
#so on. We can use three components to create the matrix ,
comp1_img=matrix(0,232,232)
i=1
for (m in 1:232)
    {comp1_img[m,]=pca_image$scores[i:(232*m),1]
    i=232*m+1
    }
comp2_img=matrix(0,232,232)
i=1
for (m in 1:232)
    {comp2_img[m,]=pca_image$scores[i:(232*m),2]
    i=232*m+1
    }
comp3_img=matrix(0,232,232)
i=1
for (m in 1:232)
    {comp3_img[m,]=pca_image$scores[i:(232*m),3]
    i=232*m+1
    }
#use component1 to reconstruct image
plot(c(0,1),c(0,1),t='n')
rasterImage((comp1_img-min(comp1_img))/(max(comp1_img)-min(comp1_img)),0,0,1,1)
```

```r
#use component2 to reconstruct image
plot(c(0,1),c(0,1),t='n')
rasterImage((comp2_img-min(comp2_img))/(max(comp2_img)-min(comp2_img)),0,0,1,1)
```

```r
#use component3 to reconstruct image
plot(c(0,1),c(0,1),t='n')
rasterImage((comp3_img-min(comp3_img))/(max(comp3_img)-min(comp3_img)),0,0,1,1)
```

```r
#we can see from the reconstructed images that the component1 gives better reconstructed image than the others. But we can use
#three components together to reconstruct the image better.
```
